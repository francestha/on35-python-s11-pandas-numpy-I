```python
coluna1 = df['UMIDADE RELATIVA DO AR, HORARIA (%)']
coluna1

# Funções para calcular o valor mínimo, máximo, soma e média

# Valor mínimo
minimo = df['UMIDADE RELATIVA DO AR, HORARIA (%)'].min()

# Valor máximo
maximo = df['UMIDADE RELATIVA DO AR, HORARIA (%)'].max()

# Soma dos valores
soma = df['UMIDADE RELATIVA DO AR, HORARIA (%)'].sum()

# Média dos valores
media = df['UMIDADE RELATIVA DO AR, HORARIA (%)'].mean()

print(f"Mínimo: {minimo}, Máximo: {maximo}, Soma: {soma}, Média: {media} da Umidade Relativa do Ar.")
```

```python
#Identifica valores nulos no DataFrame, retornando um DataFrame booleano.

# Verifica valores nulos
nulos = df.isnull()
nulos
```

```python
# Contagem de valores nulos em cada coluna
nulos_por_coluna = df.isnull().sum()
nulos_por_coluna
```

```python
# Remove linhas com qualquer valor nulo
df_sem_nulos = df.dropna()
df_sem_nulos

# Remove colunas que contenham valores nulos
df_sem_nulos_colunas = df.dropna(axis=1, how='all')
df_sem_nulos_colunas
```

```python
# Preenche valores nulos com um valor específico.

# Preenche valores nulos com 0
df_preenchido = df.fillna(0)
df_preenchido
```

```python
# Substitui todos os valores 0 por NaN
df_substituido = df.replace('0', 'NaN')
df_substituido
```
